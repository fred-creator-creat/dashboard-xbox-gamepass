```python
import pandas as pd

# 1. Carregar o arquivo completo
excel_file = pd.ExcelFile('base.xlsx')

# 2. Procurar em todas as abas qual delas tem a coluna 'Plan'
df_vendas = None
for sheet in excel_file.sheet_names:
    temp_df = pd.read_excel('base.xlsx', sheet_name=sheet)
    # Limpa os nomes das colunas para garantir
    temp_df.columns = [str(c).strip() for c in temp_df.columns]
    if 'Plan' in temp_df.columns:
        df_vendas = temp_df
        break

# 3. Verifica se encontrou e gera o arquivo
if df_vendas is not None:
    summary = df_vendas.groupby('Plan')['Total Value'].sum().reset_index()
    with pd.ExcelWriter('dados_para_dashboard.xlsx') as writer:
        df_vendas.to_excel(writer, sheet_name='Base', index=False)
        summary.to_excel(writer, sheet_name='Summary', index=False)
    print("✅ O arquivo 'dados_para_dashboard.xlsx' foi criado com sucesso.")
else:
    print("❌ Erro: Não encontrei os dados.")
```
